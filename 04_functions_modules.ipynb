## Создание функции

```python
patients = [
    {'USUBJID':"1001-01", "WEIGHT":74, "HEIGHT":1.80},
    {'USUBJID':"1001-02", "WEIGHT":81, "HEIGHT":1.73},
    {'USUBJID':"1001-03", "WEIGHT":67, "HEIGHT":1.71},
    {'USUBJID':"1001-04", "WEIGHT":90, "HEIGHT":1.83},
    {'USUBJID':"1001-05", "WEIGHT":75, "HEIGHT":1.75},
]

for patient in patients:
    patient['BMI'] = calculate_bmi(patient['WEIGHT'], patient['HEIGHT'])

print(patients)
```

```python
def calculate_average(*values):
    average = sum(values) / len(values)
    return average
```

```python
blood_test = [5.4, 6.2, 7.1, 5.8]
average_test_result = calculate_average(*blood_test)
print(f"Среднее значение анализа - {average_test_result}")
```

## Функции - значения по умолчанию

```python
def calculate_bmi(wei=75, hei=1.75):
    bmi_calculated = wei / (hei ** 2)
    return bmi_calculated

bmi = calculate_bmi(80)
print(bmi)
```

## Функции - Позиционные и именные аргументы

```python
def calculate_bmi(wei=75, hei=1.75):
    bmi_calculated = wei / (hei ** 2)
    return bmi_calculated

bmi = calculate_bmi(hei=height, wei=weight)
print(bmi)
```

```python
def calculate_bmi(wei=75, hei=1.75):
    bmi_calculated = wei / (hei ** 2)
    return bmi_calculated

bmi = calculate_bmi(weight, hei=height)
print(bmi)
```
