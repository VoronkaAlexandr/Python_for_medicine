```python
allergy_patient_count = 0
for patient in patients:
    if patient['allergy_to_penicilline'] == True:
        allergy_patient_count += 1 # allergy_patient_count = allergy_patient_count + 1

print(allergy_patient_count)
```

```python
patients = [
    {"name":"Иванов Иван", "allergy_to_penicilline":True, "treatment":"Пенициллин", "AVAL": 12.6},
    {"name":"Петров Петр", "allergy_to_penicilline":False, "treatment":"Амоксициллин", "AVAL": 2},
    {"name":"Сидоров Сидор", "allergy_to_penicilline":True, "treatment":"Пенициллин", "AVAL": 0.25},
    {"name":"Михайлова Мария", "allergy_to_penicilline":False, "treatment":"Цефтриаксон", "AVAL": 7},
]

allergy_patient_count = 0
allergy_patient_sum_aval = 0
for patient in patients:
    if patient['allergy_to_penicilline'] == True:
        allergy_patient_count += 1 # allergy_patient_count = allergy_patient_count + 1
        allergy_patient_sum_aval += patient['AVAL']

print(allergy_patient_count)
print(allergy_patient_sum_aval)

avg_aval_allergy = allergy_patient_sum_aval / allergy_patient_count
print(avg_aval_allergy)
```

## Цикл while

```python
normal_glucose_level = 6.0
glucose_level = 9.0
days = 0
```

```python
while glucose_level > normal_glucose_level:
    days += 1
    glucose_level -= 0.2 # glucose_level = glucose_level - 0.2
```
