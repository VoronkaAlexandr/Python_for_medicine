```python
class ClinicalStudyPatient:
    def __init__(self, name, age, cholesterol_level):
        self.name = name
        self.age = age
        self.cholesterol_level = cholesterol_level
    def avg_cholesterol(self):
        return sum(self.cholesterol_level) / len(self.cholesterol_level)
    def assess_risk(self):
        avg_chol = self.avg_cholesterol()
        if avg_chol > 5.0:
            return f"У пациента {self.name} высокий риск сердечно-сосудистых заболеваний"
        else:
            return f"У пациента {self.name} нормальный уровень холестерина"
    def report(self):
        avg_chol = self.avg_cholesterol()
        risk = self.assess_risk()
        return f"Пациент: {self.name}, Возраст: {self.age} лет, Средний уровень холестерина: {avg_chol:.2f}, Риск: {risk}"
```

```python
patient1 = ClinicalStudyPatient("Мария Петрова", 30, [5.6, 5.9, 6.0, 5.7, 6.3])
```

```python
patient1.report()
```

```python
patient1.assess_risk()
```

## Инкапсуляция

```python
class Patient:
    def __init__(self, name, age, diagnosis):
        self.name = name
        self._age = age
        self.__diagnosis = diagnosis

    def get_age(self):
        return self._age

    def set_diagnosis(self, new_diagnosis):
        if isinstance(new_diagnosis, str):
            self.__diagnosis = new_diagnosis
        else:
            print("Ошибка: диагноз должен быть строкой.")

    def get_diagnosis(self):
        return self.__diagnosis
```

## Полиморфизм

```python
class Patient(Person):
    def __init__(self, name, age, diagnosis):
        super().__init__(name, age)
        self.__diagnosis = diagnosis

    def get_diagnosis(self):
        return f"Диагноз - {self.__diagnosis}"
```

```python
class CriticalPatient(Patient):
    def __init__(self, name, age, diagnosis, critical_condition):
        super().__init__(name, age, diagnosis)
        self.critical_condition = critical_condition

    def get_diagnosis(self):
        return f"Критическое состояние пациента {self.name}, состояние - {self.critical_condition}, диагноз - {self._Patient__diagnosis}"
```
